```python
import ollama

def exit_llm():
    while True:
        user_response = input("\nWould you like to ask something else? (y/n): ")
        if user_response.lower() == 'n':
            return True
        elif user_response.lower() == 'y':
            return False
        else:
            print("\nPlease enter y or n")
            
def get_input():
    while True:
        user_response = input("\nHow can I help you?\n")
        
        if user_response == "":
            print("\nPlease enter a prompt")
            continue
        break
    return user_response

def get_response(prompt):
    try:
        response = ollama.chat(
            model='gemma:2b',
            messages=[
                # {'role': 'system', 'content': 'You are a helpful assistant.'},
                {'role': 'user', 'content': prompt}
            ]
        )
    except Exception as e:
        print("\nThere was an error getting an LLM response\n")
        return e
    return response['message']['content']

while True:  
    user_prompt = get_input()
    print("\nInput Prompt:", user_prompt)
    
    llm_response = get_response(user_prompt)
    print("\nResponse:", llm_response)
    
    if exit_llm():
        print("\nHope to see you again!")
        break
```
