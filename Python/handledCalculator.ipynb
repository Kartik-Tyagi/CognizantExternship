```python
#Calculator with exception handling

import time
from datetime import datetime

while True:
    print("Welcome to the Handled calculator!")
    time.sleep(1)
    print("\nPlease selective which option you would like by entering the corresponding number:")
    time.sleep(1)
    print("(1) Addition")
    time.sleep(1)
    print("(2) Subtraction")
    time.sleep(1)
    print("(3) Multiplication")
    time.sleep(1)
    print("(4) Division")
    time.sleep(1)
    print("(5) Exit")
    
    user_option = 0
    while True:
        user_response = input("Which option do you choose? ")
        try:
            user_option = int(user_response)
        except:
            print("\nPlease enter a valid number corresponding to the option you wish to select, the is the number presented like (6)")
            timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            with open("./error_log.txt", "a") as file:
                file.write(f"[{timestamp}] :- When prompting user to choose a menu option: ValueError (not entered an integer)\n")
            continue
        if user_option < 1 or user_option > 5:
            print("\nPlease enter a valid number corresponding to the option you wish to select, the is the number presented like (6)")
            timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            with open("./error_log.txt", "a") as file:
                file.write(f"[{timestamp}] :- When prompting user to choose a menu option: ValueError (Entered an integer but not between 1 and 5)\n")
            continue
        break
    
    if user_option == 5:
        print("\nHopefully I see you again!!")
        break
    
    user_number1 = 0
    user_number2 = 0
    
    while True:
        user_response = input("What's your first number: ")

        try:
            user_number1 = int(user_response)
        except:
            print("\nGive an integer please")
            timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            with open("./error_log.txt", "a") as file:
                file.write(f"[{timestamp}] :- When prompting user to enter the first number: ValueError (not entered an integer)\n")
            continue
        break

    while True:
        user_response = input("Be prepared for awesomeness, what's your second number: ")

        try:
            user_number2 = int(user_response)
        except:
            print("\nGive an integer please")
            timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            with open("./error_log.txt", "a") as file:
                file.write(f"[{timestamp}] :- When prompting user to enter the second number: ValueError (not entered an integer)\n")
            continue
        if user_option == 4 and user_number2 == 0:
            print("\nSeems like you want to divide, this integer cannot be 0")
            timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            with open("./error_log.txt", "a") as file:
                file.write(f"[{timestamp}] :- When prompting user to enter the second number: ZeroDivisionError (Entered a 0 when user wishes to divide)\n")
            continue
        break
    
    time.sleep(1)
    
    if user_option == 1:
        print("\nLet's see if you can handle this 5-star addition")
        time.sleep(1)
        print(user_number1, "+", user_number2, "=", user_number1+user_number2)
    elif user_option == 2:
        print("\nLet's see if you can handle this legend level subtraction")
        time.sleep(1)
        print(user_number1, "-", user_number2, "=", user_number1-user_number2)
    elif user_option == 3:
        print("\nLet's see if you can handle this godly multiplication")
        time.sleep(1)
        print(user_number1, "x", user_number2, "=", user_number1*user_number2)
    else:
        print("\nLet's see if you can handle this fully robust division")
        time.sleep(1)
        print(user_number1, "/", user_number2, "=", user_number1/user_number2)
    print("")
    time.sleep(1.5)
```
