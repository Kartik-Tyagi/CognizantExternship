```python
#Number Guessing Game

import random
import time

number_to_guess = random.randint(1, 100)

print("Instructions:")
print("I have thought of a number between 1 and 100")
print("You get 10 tries to guess the number")
print("If you are off by 10 or more, I will reply with either too high or too low")
print("If your guess is within a difference of 10 from my number, I will tell you that you are close")
print("Happy guessing!!!")

time.sleep(5)
attempts = 0

while attempts < 10:
    user_input = input(f"What's your guess? (Attempts left: {10 - attempts}): ")
    user_number = 0
    attempts += 1
    try:
        user_number = int(user_input)
    except:
        print("You are wasting guesses, enter a integer number between 1 and 100 next time")
        continue
    if user_number > 100 or user_number < 1:
        print("You are wasting guesses, enter a integer number between 1 and 100 next time")
        continue
    if user_number == number_to_guess:
        print("WOAHHHH!! YOU GOT IT!!")
        print("Attempts taken:", attempts)
    elif user_number > number_to_guess:
        if user_number - number_to_guess < 10:
            print("Oooooooh, so close! You guessed a little high")
        else:
            print("Too Highhhh. Try again")
        continue
    else:
        if number_to_guess - user_number < 10:
            print("Oooooooh, so close! You guessed a little low")
        else:
            print("Too Lowwww. Try again")
        continue
    
    time.sleep(2)
    attempts = 0
    user_response = input("If you want to play again, enter p: ")
    if user_response == 'p':
        number_to_guess = random.randint(1, 100)
        continue
    else:
        break

if attempts == 10:
    print("Better luck next time! Attempts left: 0")
```
