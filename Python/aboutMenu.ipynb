```python
#Creating an about menu

import time
import turtle

def factorial(n):
    if n == 0:
        return 1
    if n == 1:
        return 1
    return n*factorial(n-1)

def fibonacci(n):
    if n == 1:
        return 1
    if n == 2:
        return 1
    return fibonacci(n-1) + fibonacci(n-2)

def curve(t, length, depth):
    if depth == 0:
        t.forward(length)
    else:
        length /= 3.0
        curve(t, length, depth - 1)
        t.left(60)
        curve(t, length, depth - 1)
        t.right(120)
        curve(t, length, depth - 1)
        t.left(60)
        curve(t, length, depth - 1)

while True:
    print("Welcome to the recursive calculator!")
    time.sleep(1)
    print("\nPlease selective which option you would like by entering the corresponding number:")
    time.sleep(1)
    print("(1) Calculate the factorial of a number")
    time.sleep(1)
    print("(2) Find the nth Fibonacci number")
    time.sleep(1)
    print("(3) Draw a recursive fractal pattern")
    time.sleep(1)
    print("(4) Exit")
    
    user_option = 0
    while True:
        user_response = input("Which option do you choose")
        try:
            user_option = int(user_response)
        except:
            print("\nPlease enter a valid number corresponding to the option you wish to select, the is the number presented like (5)")
            continue
        if user_option < 1 or user_option > 4:
            print("\nPlease enter a valid number corresponding to the option you wish to select, the is the number presented like (5)")
            continue
        break
    
    time.sleep(1)
    
    if user_option == 1:
        print("\n\nSeems like you have chosen our factorial option. Great Choice!\n")
        user_number = 0
        while True:
            user_response = input("Which number's factorial would you like to calculate? ")
            try:
                user_number = int(user_response)
            except:
                print("\nPlease enter an integer greater than or equal to 0")
                continue
            if user_number < 0:
                print("\nPlease enter an integer greater than or equal to 0")
                continue
            break
        time.sleep(1)
        print("As you requested, the factorial of", user_number, "is:")
        print(factorial(user_number))
        
    elif user_option == 4:
        print("\n\nAwww, you are going? Hope to see you again soon")
        break
    
    elif user_option == 2:
        print("\n\nSeems like you have chosen our Fibonacci option. Classic!\n")
        user_number = 0
        while True:
            user_response = input("Which numbered term in the Fibonacci sequence would you like to know? ")
            try:
                user_number = int(user_response)
            except:
                print("\nPlease enter an integer greater than or equal to 1")
                continue
            if user_number < 1:
                print("\nPlease enter an integer greater than or equal to 1")
                continue
            break
        time.sleep(1)
        if user_number%10 == 1:
            print(f"As you requested, the {user_number}st term of the Fibonacci sequence is:")
            print(fibonacci(user_number))
        elif user_number%10 == 2:
            print(f"As you requested, the {user_number}nd term of the Fibonacci sequence is:")
            print(fibonacci(user_number))
        elif user_number%10 == 3:
            print(f"As you requested, the {user_number}rd term of the Fibonacci sequence is:")
            print(fibonacci(user_number))
        else:
            print(f"As you requested, the {user_number}th term of the Fibonacci sequence is:")
            print(fibonacci(user_number))
            
    else:
        print("\n\nOooooh, fancy! Get ready for our recursive pattern")
        screen = turtle.Screen()
        screen.bgcolor("white")
        t = turtle.Turtle()
        t.speed("fastest")
        t.penup()
        t.goto(-150, 90)
        t.pendown()
        user_number = 0
        while True:
            user_response = input("Which numbered term in the Fibonacci sequence would you like to know? ")
            try:
                user_number = int(user_response)
            except:
                print("\nPlease enter an integer greater than or equal to 1")
                continue
            if user_number < 1:
                print("\nPlease enter an integer greater than or equal to 1")
                continue
            break        
        for i in range(user_number):
            curve(t, 300, 4)
            t.right(120)
        screen.exitonclick()
    
    print("\n")
```
